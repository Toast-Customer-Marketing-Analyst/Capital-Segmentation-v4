```python
Leads_df = df.loc[(df['IS_LEAD'] == True) & (df['IS_POST_RAMP'] == True)]
MQLs_df = df.loc[(df['IS_MQL'] == True) & (df['IS_POST_RAMP'] == True)]
Opps_df = df.loc[(df['IS_OPP'] == True) & (df['IS_POST_RAMP'] == True)]
Deals_df = df.loc[(df['IS_DEAL'] == True) & (df['IS_POST_RAMP'] == True)]
Declined_df = df.loc[(df['IS_DECLINED'] == True) & (df['IS_POST_RAMP'] == True)]
```

```python
#this function builds a df showing the key capital profile metrics of Cx that share an attribute (passed in as the "grouping" argument)

def function(all_df, leads_df, mqls_df, opps_df, deals_df, declined_df, grouping):
    capital_profile = all_df.groupby(grouping).agg({'SALESFORCE_ACCOUNTID': 'count'})
    capital_profile.columns = ['Cx_Count']
    capital_profile['Leads_Count'] = leads_df.groupby(grouping).agg({'SALESFORCE_ACCOUNTID': 'count'})
    capital_profile['MQLs_Count'] = mqls_df.groupby(grouping).agg({'SALESFORCE_ACCOUNTID': 'count'})
    capital_profile['Opps_Count'] = opps_df.groupby(grouping).agg({'SALESFORCE_ACCOUNTID': 'count'})
    capital_profile['Deals_Count'] = deals_df.groupby(grouping).agg({'SALESFORCE_ACCOUNTID': 'count'})
    capital_profile['Declined_Count'] = declined_df.groupby(grouping).agg({'SALESFORCE_ACCOUNTID': 'count'})
    
    capital_profile['Cx_to_Lead'] = np.round((capital_profile['Leads_Count']/capital_profile['Cx_Count']), decimals=3)
    capital_profile['Lead_to_MQL'] = np.round((capital_profile['MQLs_Count']/capital_profile['Leads_Count']), decimals=3)
    capital_profile['MQL_to_Opp'] = np.round((capital_profile['Opps_Count']/capital_profile['MQLs_Count']), decimals=3)
    capital_profile['Opp_to_Deal'] = np.round((capital_profile['Deals_Count']/capital_profile['Opps_Count']), decimals=3)
    capital_profile['Lead_to_Deal'] = np.round((capital_profile['Deals_Count']/capital_profile['Leads_Count']), decimals=3)
    capital_profile['MQL_to_Declined'] = np.round((capital_profile['Declined_Count']/capital_profile['MQLs_Count']), decimals=3)
    capital_profile.reset_index(inplace=True)
      
    return capital_profile
```

```python
def new_function(all_df, leads_df, mqls_df, opps_df, deals_df, declined_df, col1, col2=None):
    capital_profile = all_df.groupby([col1,col2]).agg({'SALESFORCE_PARENT_ACCOUNTID': 'count'})
    capital_profile.columns = ['Cx_Count']
    capital_profile['Leads_Count'] = leads_df.groupby([col1,col2]).agg({'SALESFORCE_ACCOUNTID': 'count'})
    capital_profile['MQLs_Count'] = mqls_df.groupby([col1,col2]).agg({'SALESFORCE_ACCOUNTID': 'count'})
    capital_profile['Opps_Count'] = opps_df.groupby([col1,col2]).agg({'SALESFORCE_ACCOUNTID': 'count'})
    capital_profile['Deals_Count'] = deals_df.groupby([col1,col2]).agg({'SALESFORCE_ACCOUNTID': 'count'})
    capital_profile['Declined_Count'] = declined_df.groupby([col1,col2]).agg({'SALESFORCE_ACCOUNTID': 'count'})
    
    capital_profile['Cx_to_Lead'] = np.round((capital_profile['Leads_Count']/capital_profile['Cx_Count']), decimals=3)
    capital_profile['Lead_to_MQL'] = np.round((capital_profile['MQLs_Count']/capital_profile['Leads_Count']), decimals=3)
    capital_profile['MQL_to_Opp'] = np.round((capital_profile['Opps_Count']/capital_profile['MQLs_Count']), decimals=3)
    capital_profile['MQL_to_Deal'] = np.round((capital_profile['Deals_Count']/capital_profile['MQLs_Count']), decimals=3)
    capital_profile['Opp_to_Deal'] = np.round((capital_profile['Deals_Count']/capital_profile['Opps_Count']), decimals=3)
    capital_profile['Lead_to_Deal'] = np.round((capital_profile['Deals_Count']/capital_profile['Leads_Count']), decimals=3)
    capital_profile['MQL_to_Declined'] = np.round((capital_profile['Declined_Count']/capital_profile['MQLs_Count']), decimals=3)
    capital_profile.reset_index(inplace=True)
      
    return capital_profile
```

```python
def plot_function(df, dimension, measure, degrees, series=None, dimension2=None):
  df = sns.barplot(data=df, x=dimension, hue=dimension2, y=measure, order=series)
  df.set_xticklabels(df.get_xticklabels(), rotation=degrees)
  for p in df.patches: 
    df.annotate(np.round(p.get_height(),decimals=2), (p.get_x()+p.get_width()/2., p.get_height()), ha='left', va='center', xytext=(0, 10), textcoords='offset points')
    #plt.xticks([0,1,2,3,4], series)
  sns.set(rc={'figure.figsize':(11.7,8.27)})
  sns.set_style("white")
```

```python
# this function builds a subplot with 4 barcharts where each plot corresponds to one profiling measure

def subplot_function(df, data, x1, align, degrees, series=None, title=None, ax1_title=None, ax2_title=None, ax3_title=None, ax4_title=None, x2=None):
  fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
  fig.suptitle(title)
  ax1.set_facecolor('white')
  ax2.set_facecolor('white')
  ax3.set_facecolor('white')
  ax4.set_facecolor('white') 
  df = sns.barplot(  y='Lead_to_MQL', x=x1, hue=x2, data=data,  orient='v' , ax=ax1, order=series, ci="sd")
  for p in df.patches: 
    df.annotate(np.round(p.get_height(),decimals=3), (p.get_x()+p.get_width()/2., p.get_height()), ha=align, va='center', xytext=(0, 10), textcoords='offset points')
    df.set_xticklabels(df.get_xticklabels(), rotation=degrees)
  df = sns.barplot(  y="Lead_to_Deal", x=x1, hue=x2, data=data,  orient='v' , ax=ax2, order=series, ci="sd")
  for p in df.patches: 
    df.annotate(np.round(p.get_height(),decimals=3), (p.get_x()+p.get_width()/2., p.get_height()), ha=align, va='center', xytext=(0, 10), textcoords='offset points')
    df.set_xticklabels(df.get_xticklabels(), rotation=degrees)
  df = sns.barplot(  y='Opp_to_Deal', x=x1, hue=x2, data=data,  orient='v' , ax=ax3, order=series, ci="sd")
  for p in df.patches: 
    df.annotate(np.round(p.get_height(),decimals=3), (p.get_x()+p.get_width()/2., p.get_height()), ha=align, va='center', xytext=(0, 10), textcoords='offset points')
    df.set_xticklabels(df.get_xticklabels(), rotation=degrees)
  df = sns.barplot(  y="MQL_to_Declined", x=x1, hue=x2, data=data,  orient='v' , ax=ax4, order=series, ci="sd")
  for p in df.patches: 
    df.annotate(np.round(p.get_height(),decimals=3), (p.get_x()+p.get_width()/2., p.get_height()), ha=align, va='center', xytext=(0, 10), textcoords='offset points')
    df.set_xticklabels(df.get_xticklabels(), rotation=degrees)
    #plt.xticks([0,1,2,3,4])
  ax1.set_title(ax1_title)
  ax2.set_title(ax2_title)
  ax3.set_title(ax3_title)
  ax4.set_title(ax4_title) 
  #ax1.get_xaxis().set_visible(False)
  #ax2.get_xaxis().set_visible(False)
  plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.3)
```
